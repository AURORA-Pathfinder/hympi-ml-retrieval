```python
geometry = [Point(lon, lat) for lon, lat in zip(mh_data[:,1], mh_data[:,0])]
mh_times = mh_data[:,-1].astype(int).astype(str)
mh_times = [x[-4:] for x in mh_times]

gdf = gpd.GeoDataFrame(geometry=geometry)

gdf['times_npz'] = mh_times
gdf['index_npz'] = list(range(len(mh_times)))
gdf.set_crs(epsg=4326, inplace=True)
gdf.tail()
```

```python
gdf_all = gdf.to_crs(epsg=6933)
gdf_cpl = df.to_crs(epsg=6933)
```

```python
gdf_all.head()
```

```python
gdf_cpl.head()
```

```python
merged_df = []

for i in set(gdf_cpl["time"]):
    cpl_at_time = gdf_cpl[gdf_cpl['time'] == i]
    npz_at_time = gdf_all[gdf_all['times_npz'] == i]
    cpl_at_time.sjoin_nearest(npz_at_time, how='left', distance_col='distance')
    merged_df.append(cpl_at_time)
merged_df = pd.concat(merged_df, ignore_index=True)
merged_df.head()
```

```python
gdf_merged = gpd.sjoin_nearest(gdf_cpl, gdf_all, how='left', distance_col='distances')


```
